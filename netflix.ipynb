```python
import matplotlib.pyplot as plt

numerical_features = ['release_year', 'duration_minutes']
numerical_features_to_plot = [col for col in numerical_features if col != 'date_added_year']

plt.figure(figsize=(10, 4))
for i, col in enumerate(numerical_features_to_plot):
    plt.subplot(1, 2, i + 1)
    plt.boxplot(df_model[col], vert=False, patch_artist=True, showfliers=True, boxprops=dict(facecolor='lightblue'))
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```
